I removed outliers that out of the range of mean value plus or reduce 3 plus standard deviation.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

LR = LinearRegression()
y=df['PE'].to_numpy().reshape(-1, 1)
univariate_regression_coefficients=[]

for key in df:
    if key!='PE':
        print('predictor: ', key)
        x=df[key].to_numpy()
        print('removing outliers...')
        for i in range(len(x)):
            if abs(x[i]-x.mean())<=(3*x.std()):
                continue

            elif x[i]>x.mean()+(3*x.std()):
                x[i]=x.mean()+(3*x.std())
            else:
                x[i]=x.mean()-(3*x.std())
            # x=x[abs(x-x.mean())<=(3*x.std())]
        x=x.reshape(-1, 1)

        LR.fit(x, y)
        predictions=LR.predict(x)

        X = sm.add_constant(x)
        model = sm.OLS(y, X)
        results=model.fit()
        predictions2 = results.fittedvalues
        print(results.summary())
        univariate_regression_coefficients.append(results.params[1])

        plt.scatter(x , y ,color ='green')
        plt.plot(x ,predictions ,color='blue',linewidth =3)
        plt.plot(x ,predictions2 ,color='red',linewidth =3)
        plt.show()
```

(i) KNN Regression:
i. Perform k-nearest neighbor regression for this dataset using both normalized
and raw features. Find the value of k âˆˆ {1, 2, . . . , 100} that gives you the
best fit. Plot the train and test errors in terms of 1/k.

```python
from tqdm import tqdm
from sklearn.neighbors import KNeighborsRegressor as knn
train_mse, test_mse = [], []
print('Perform k-nearest neighbor regression using raw features:')
for k in tqdm(range(1,101)):
    model_knn1=knn(k)
    model_knn1.fit(x_train,y_train)
    pred_train = model_knn1.predict(x_train)
    pred_test = model_knn1.predict(x_test)
    train_mse.append(metrics.mean_squared_error(y_train,pred_train))
    test_mse.append(metrics.mean_squared_error(y_test,pred_test))
print('MSE when k = ', test_mse.index(min(test_mse))+1, ' is best fitted (raw):',min(test_mse))
plt.plot([1/k for k in range(1,101)], test_mse, label='test_mse')
plt.plot([1/k for k in range(1,101)],train_mse,color='red', label='train_mse')
plt.legend()
plt.show()

print('Perform k-nearest neighbor regression using normalized features:')
train_test = np.concatenate([x_train, x_test], axis=0)
train_len = np.shape(x_train)[0]
train_test = train_test/train_test.max(axis=0)
x_train_normalized = train_test[:train_len]
x_test_normalized = train_test[train_len:]

train_mse_normalized, test_mse_normalized = [], []
for k in tqdm(range(1,101)):
    model_knn2=knn(k)
    model_knn2.fit(x_train_normalized,y_train)
    pred_train = model_knn2.predict(x_train_normalized)
    pred_test = model_knn2.predict(x_test_normalized)
    train_mse_normalized.append(metrics.mean_squared_error(y_train,pred_train))
    test_mse_normalized.append(metrics.mean_squared_error(y_test,pred_test))
min_mse = min(test_mse_normalized)
print('MSE when k = ', test_mse_normalized.index(min_mse)+1, ' is best fitted (normalized):', min_mse)
plt.plot([1/k for k in range(1,101)], test_mse_normalized, label='test_mse')
plt.plot([1/k for k in range(1,101)],train_mse_normalized,color='red', label='train_mse')
plt.legend()
plt.show()
```
